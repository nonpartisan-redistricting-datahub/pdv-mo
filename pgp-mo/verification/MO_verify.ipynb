### Replication of PGP's MO Shapefile

The final reports for both PGP's final file and the replication file are the same. Both received a [vote score](https://github.com/OpenPrecincts/verification#vote-score) of 0.99, an excellent score.

[worst_county_vote_score](https://github.com/OpenPrecincts/verification#vote-score) is also very good at 0.99. This suggests the county with the most deviant vote score is only 1 percentage point from the ideal value of 1. 

[county_vote_score_dispersion](https://github.com/OpenPrecincts/verification#county-vote-score-dispersion) is low at 5.43 with an ideal of 0. This indicates that there is a not a substantial difference between the replicated file and PGP's file with regards to offical state election data records at the county-level.

[median_county_area_difference_score](https://github.com/OpenPrecincts/verification#area-difference-score) is at the ideal score of 0.
